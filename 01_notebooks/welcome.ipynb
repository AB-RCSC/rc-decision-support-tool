```python
# Create an Output widget for the pop-up window content
output_popup = widgets.Output()

# Function to show the pop-up window
def show_popup(b):
    with output_popup:
        output_popup.clear_output()
        display(popup_window)

# Function to close the pop-up window
def close_popup(b):
    output_popup.clear_output()

# Create a close button for the pop-up window
close_button = widgets.Button(
    description="Close",
    layout=widgets.Layout(width='100px')
)
# Calling the closing function when close button is clicked
close_button.on_click(close_popup)
```

```python
# Create the pop-up window container
popup_window = widgets.VBox(
    children=[
        widgets.HTML(iframe),
        close_button
    ],
    layout=widgets.Layout(
        display='flex',
        flex_flow='column',
        align_items='center',
        justify_content='center',
#         border='solid 2px grey',
        padding='5px',
        width='auto',
        height='auto'
    )
)
```

```python
# Create button for pop up
button = widgets.Button(
    description="Click to know more", # ALIF - CHANGED FROM "Click"
    layout=widgets.Layout(width='auto')
)

button.on_click(show_popup)
```

```python
# Create a buttons for submitting the question
submit_button = widgets.Button(description="Start >>", layout=widgets.Layout(border='1px solid black'))
```

```python
# blank widget used at the left corner to position the "submit" button at the right corner
blank_widget = widgets.HTML(value='')
```

```python
output = widgets.Output()
def on_button_clicked(event):
    with output:
        clear_output()
        next_notebook = "objective"
            
        notebook_filename = next_notebook + ".ipynb"
        # JavaScript to open the URL
        js_code = f"window.open('{notebook_filename}', '_self');"
        back_list.append("welcome")
        with open (back_path, "w")as f:
            json.dump(back_list, f)

        display(Javascript(js_code))

submit_button.on_click(on_button_clicked)
```
